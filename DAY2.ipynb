```python
# P1=38.P2=30,P3=25
#fill the ages
def sri(col):
  Age=col[0]
  Pclass=col[1]
  if pd.isnull(Age):
    if Pclass==1:
      return 38
    elif Pclass==2:
      return 30
    else:
      return 25
  else:
    return Age
```

```python
titanic['Age']=titanic[['Age','Pclass']].apply(sri,axis=1)
```

```python
import torch
from pytorch_transformers import GPT2Tokenizer,GPT2LMHeadModel
data="An Apple a day keeps the doctor away"
tokenizer=GPT2Tokenizer.from_pretrained("gpt2")
index_token=tokenizer.encode(data)
tokens_tensor=torch.tensor([index_token])
tokens_tensor
```

```python
model=GPT2LMHeadModel.from_pretrained("gpt2")
model.eval()
```

```python
tokens_tensor=tokens_tensor.to("cuda")
model.to("cuda")
```

```python
with torch.no_grad():
  output=model(tokens_tensor)
  #above output is kept in a model and storing the output in a var
prediction=output[0]
prediction
```

```python
pi=torch.argmax(prediction[0,-1,:]).item()
pt=tokenizer.decode(index_token+[pi])
```
